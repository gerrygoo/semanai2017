```python
i = 0

for s in text1.tokens:
        
    if not isWord(s):
        continue
    
    print('word = ' + s)
    print('popularity = ' + str(getPopularity(text1.count(s), len(text1)))  + "%")
    print()
    i = i + 1
    
    if i > 100:
        break
```
